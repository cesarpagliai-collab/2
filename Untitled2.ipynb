```python
import streamlit as st
import pandas as pd
import time
import json
from datetime import datetime

# --- 1. CLASES BASE (MODELOS DE DATOS) ---
# (Estas clases no cambian)

class PLADECO:
    def __init__(self, nombre_comuna, periodo, lineamientos):
        self.nombre_comuna = nombre_comuna
        self.periodo = periodo
        self.lineamientos = lineamientos
        self.enfoques = []
        self.marco_seguimiento = {}

class TeoriaDeCambio:
    def __init__(self, nombre, objetivo_pladeco, problema):
        self.nombre = nombre
        self.version = 1.0
        self.objetivo_pladeco = objetivo_pladeco
        self.problema = problema
        self.cadena_causal = {
            "impacto": objetivo_pladeco,
            "resultados": [], "productos": [], "actividades": []
        }
        self.supuestos = []

class Proyecto:
    def __init__(self, nombre, lider):
        self.nombre = nombre
        self.lider = lider
        self.estado = "Planificado"
        self.progreso = 0
        self.presupuesto_asignado = 10000 # Simulación
        self.presupuesto_gastado = 0
        self.actividades_productos = []

class Portafolio:
    def __init__(self, nombre, tdc_asociada):
        self.nombre = nombre
        self.tdc_asociada = tdc_asociada
        self.proyectos = []
        self.riesgos = []

class Indicador:
    def __init__(self, descripcion, resultado_asociado):
        self.descripcion = descripcion
        self.resultado_asociado = resultado_asociado
        self.linea_base = 0
        self.meta = 100
        self.valor_actual = 0
        self.fuente_datos = "Manual"

    @property
    def avance(self):
        if self.meta == self.linea_base:
            return 100.0 if self.valor_actual >= self.meta else 0.0
        return max(0, min(100, ((self.valor_actual - self.linea_base) / (self.meta - self.linea_base)) * 100))

# --- 2. LÓGICA DE LOS MÓDULOS ADAPTADA PARA STREAMLIT ---
# (Funciones que operan sobre st.session_state)

def vincular_pladeco_logica():
    lineamientos = {
        "Desarrollo Económico Sostenible": ["Aumentar empleabilidad juvenil", "Fomentar el turismo local"],
        "Cohesión Social y Calidad de Vida": ["Mejorar espacios públicos", "Reducir la delincuencia"]
    }
    pladeco_obj = PLADECO("Villa Esperanza", "2024-2028", lineamientos)
    pladeco_obj.enfoques = ["Género", "Desarrollo Sostenible", "Participación Ciudadana"]
    pladeco_obj.marco_seguimiento = {
        "descripcion": "Se mencionan cuentas públicas anuales y reportes semestrales.",
        "calificacion": "Simple", "niveles": ["No Existente", "Simple", "Estándar", "Avanzado"]
    }
    st.session_state.pladeco = pladeco_obj

def crear_tdc_logica(nombre, objetivo, problema):
    tdc_obj = TeoriaDeCambio(nombre, objetivo, problema)
    tdc_obj.cadena_causal["resultados"] = [{"id": "R1", "desc": "Jóvenes tienen mejores habilidades", "tipo": "convencional"}]
    tdc_obj.cadena_causal["productos"] = [{"id": "P1", "desc": "Jóvenes con CV mejorado"}]
    tdc_obj.cadena_causal["actividades"] = [{"id": "A1", "desc": "Realizar talleres de habilidades"}]
    tdc_obj.supuestos = [{"de": "A1", "a": "P1", "desc": "Los jóvenes asisten a los talleres"}]
    st.session_state.tdc = tdc_obj

def crear_portafolio_logica(nombre):
    portafolio_obj = Portafolio(nombre, st.session_state.tdc)
    for supuesto in st.session_state.tdc.supuestos:
        riesgo = {"riesgo": f"Fallo en: '{supuesto['desc']}'", "prob": "Media", "impacto": "Alto", "estado": "Activo"}
        portafolio_obj.riesgos.append(riesgo)

    p1 = Proyecto("Proyecto Capacitación", "Ana Fuentes")
    p1.actividades_productos = ["A1", "P1"]
    p1.progreso = 60
    p1.presupuesto_gastado = 4500

    portafolio_obj.proyectos.append(p1)
    st.session_state.portafolio = portafolio_obj

def configurar_indicadores_logica():
    indicadores = []
    for res in st.session_state.portafolio.tdc_asociada.cadena_causal["resultados"]:
        ind = Indicador(f"Indicador para '{res['desc']}'", res['id'])
        if "habilidades" in ind.descripcion:
             ind.valor_actual = 75 # Simular progreso
        indicadores.append(ind)
    st.session_state.indicadores = indicadores


# --- 3. INICIALIZACIÓN DE LA APLICACIÓN Y ESTADO DE SESIÓN ---

st.set_page_config(layout="wide", page_title="App Municipal de Gestión Estratégica")

# Inicializar el estado de la sesión para todos nuestros objetos
if 'pladeco' not in st.session_state:
    st.session_state.pladeco = None
if 'tdc' not in st.session_state:
    st.session_state.tdc = None
if 'portafolio' not in st.session_state:
    st.session_state.portafolio = None
if 'indicadores' not in st.session_state:
    st.session_state.indicadores = []

# --- 4. BARRA LATERAL DE NAVEGACIÓN ---

st.sidebar.title("Navegación de Módulos")

# La navegación cambia dependiendo de lo que ya se ha configurado
if st.session_state.pladeco is None:
    st.sidebar.info("Cargue un PLADECO para comenzar.")
    pagina_actual = "Módulo 1"
else:
    modulos_disponibles = ["Módulo 1: PLADECO"]
    if st.session_state.pladeco:
        modulos_disponibles.append("Módulo 2: Teoría de Cambio")
    if st.session_state.tdc:
        modulos_disponibles.append("Módulo 3: Portafolio")
    if st.session_state.portafolio:
        modulos_disponibles.append("Módulo 4: MEL")

    pagina_actual = st.sidebar.radio("Seleccione un Módulo", modulos_disponibles)


# --- 5. RENDERIZADO DE PÁGINAS (CONTENIDO PRINCIPAL) ---

# --- MÓDULO 1: VINCULACIÓN Y DASHBOARD PLADECO ---
if pagina_actual == "Módulo 1: PLADECO":
    if st.session_state.pladeco is None:
        st.title("Módulo 1: Bienvenido")
        uploaded_file = st.file_uploader("Cargue el documento PDF de su PLADECO.", type="pdf")
        if uploaded_file is not None:
            with st.spinner("Analizando PLADECO..."):
                time.sleep(2)
                vincular_pladeco_logica()
            st.success("¡PLADECO vinculado!")
            st.info("La aplicación se recargará.")
            time.sleep(1)
            st.experimental_rerun()
    else:
        pladeco = st.session_state.pladeco
        st.title(f"Panel de Control: PLADECO {pladeco.nombre_comuna}")
        # (El resto de la UI del Módulo 1 que ya teníamos)
        st.markdown("---")
        col1, col2 = st.columns(2)
        with col1:
            st.subheader("🎯 Lineamientos Estratégicos")
            for lin, obj in pladeco.lineamientos.items():
                with st.expander(f"**{lin}**"):
                    st.write(obj)
        with col2:
            st.subheader("🔍 Análisis del Marco de Seguimiento")
            st.write(pladeco.marco_seguimiento['descripcion'])
            st.select_slider('Nivel de Madurez', options=pladeco.marco_seguimiento['niveles'], value=pladeco.marco_seguimiento['calificacion'], disabled=True)
        if st.sidebar.button("⬅️ Cargar otro PLADECO"):
            # Resetear todo el estado para empezar de nuevo
            for key in st.session_state.keys():
                del st.session_state[key]
            st.experimental_rerun()


# --- MÓDULO 2: DISEÑO DE TdC ---
elif pagina_actual == "Módulo 2: Teoría de Cambio":
    st.title(pagina_actual)
    if st.session_state.tdc is None:
        st.info("Cree una nueva Teoría de Cambio para continuar.")
        objetivos = [obj for sublist in st.session_state.pladeco.lineamientos.values() for obj in sublist]
        with st.form("crear_tdc_form"):
            nombre_tdc = st.text_input("Nombre de la Teoría de Cambio", "TdC para Empleabilidad Juvenil")
            objetivo_seleccionado = st.selectbox("Seleccione el Objetivo Estratégico del PLADECO a abordar:", objetivos)
            problema = st.text_area("Describa el problema central a resolver:", "Jóvenes carecen de experiencia y redes de contacto.")
            submitted = st.form_submit_button("Crear TdC")
            if submitted:
                with st.spinner("Creando TdC..."):
                    crear_tdc_logica(nombre_tdc, objetivo_seleccionado, problema)
                st.success("TdC creada. Recargando...")
                time.sleep(1)
                st.experimental_rerun()
    else:
        tdc = st.session_state.tdc
        st.header(f"Lienzo de la TdC: '{tdc.nombre}'")
        st.write(f"**Impacto Deseado (Objetivo PLADECO):** {tdc.objetivo_pladeco}")
        st.write(f"**Problema a Resolver:** {tdc.problema}")

        st.subheader("Cadena Causal")
        # Usamos st.graphviz_chart para una visualización simple
        diagrama = f"""
        digraph tdc {{
            rankdir=BT;
            node [shape=box, style=rounded];
            impacto [label="{tdc.cadena_causal['impacto']}"];
            subgraph cluster_resultados {{
                label = "Resultados";
                R1 [label="{tdc.cadena_causal['resultados'][0]['desc']}"];
            }}
            subgraph cluster_productos {{
                label = "Productos";
                P1 [label="{tdc.cadena_causal['productos'][0]['desc']}"];
            }}
            subgraph cluster_actividades {{
                label = "Actividades";
                A1 [label="{tdc.cadena_causal['actividades'][0]['desc']}"];
            }}
            A1 -> P1 [label="{tdc.supuestos[0]['desc']}"];
            P1 -> R1;
            R1 -> impacto;
        }}
        """
        st.graphviz_chart(diagrama)


# --- MÓDULO 3: GESTIÓN DE PORTAFOLIO ---
elif pagina_actual == "Módulo 3: Portafolio":
    st.title(pagina_actual)
    if st.session_state.portafolio is None:
        st.info("Cree un nuevo Portafolio de Proyectos para continuar.")
        with st.form("crear_portafolio_form"):
            nombre_portafolio = st.text_input("Nombre del Portafolio", f"Portafolio: {st.session_state.tdc.nombre}")
            submitted = st.form_submit_button("Crear Portafolio")
            if submitted:
                with st.spinner("Creando Portafolio..."):
                    crear_portafolio_logica(nombre_portafolio)
                st.success("Portafolio creado. Recargando...")
                time.sleep(1)
                st.experimental_rerun()
    else:
        portafolio = st.session_state.portafolio
        st.header(f"Panel de Control del Portafolio: '{portafolio.nombre}'")

        st.subheader("Proyectos del Portafolio")
        for p in portafolio.proyectos:
            st.progress(p.progreso, text=f"{p.nombre} - Lider: {p.lider}")
            col1, col2 = st.columns(2)
            col1.metric("Presupuesto Gastado", f"${p.presupuesto_gastado}", f"de ${p.presupuesto_asignado}")
            col2.metric("Estado", p.estado)

        st.subheader("Registro de Riesgos (desde Supuestos de la TdC)")
        df_riesgos = pd.DataFrame(portafolio.riesgos)
        st.dataframe(df_riesgos)


# --- MÓDULO 4: MONITOREO, EVALUACIÓN Y APRENDIZAJE ---
elif pagina_actual == "Módulo 4: MEL":
    st.title(pagina_actual)

    # Configurar indicadores si no se ha hecho
    if not st.session_state.indicadores:
        configurar_indicadores_logica()

    indicadores = st.session_state.indicadores
    portafolio = st.session_state.portafolio

    st.header("Panel de Control MEL")

    st.subheader("Avance de Indicadores de Resultado (GbR)")
    for ind in indicadores:
        st.slider(ind.descripcion, min_value=ind.linea_base, max_value=ind.meta, value=ind.valor_actual, disabled=True)
        st.write(f"Avance: {ind.avance:.1f}%")

    st.subheader("Lienzo de la TdC Dinámico (Simulación)")
    # Re-usamos el diagrama de Graphviz y lo coloreamos
    # En una app real, la lógica de coloreado sería más compleja
    diagrama_mel = f"""
    digraph tdc_mel {{
        rankdir=BT;
        node [shape=box, style=rounded];
        impacto [label="{portafolio.tdc_asociada.cadena_causal['impacto']}"];
        R1 [label="{indicadores[0].descripcion}\nAvance: {indicadores[0].avance:.1f}%", style=filled, fillcolor=lightblue];
        P1 [label="{portafolio.tdc_asociada.cadena_causal['productos'][0]['desc']}", style=filled, fillcolor=lightgreen];
        A1 [label="{portafolio.tdc_asociada.cadena_causal['actividades'][0]['desc']}", style=filled, fillcolor=lightgreen];

        A1 -> P1 [label="Supuesto: {portafolio.riesgos[0]['riesgo']}", color=red, fontcolor=red];
        P1 -> R1;
        R1 -> impacto;
    }}
    """
    st.graphviz_chart(diagrama_mel)
    st.caption("Verde: Completado. Azul: En Progreso. Rojo: Riesgo identificado.")

    st.subheader("Generador de Reportes")
    audiencia = st.selectbox("Seleccione la Audiencia del Reporte", ["Equipo Interno", "Concejo Municipal", "Ciudadanía"])
    if st.button("Generar Reporte"):
        with st.spinner("Generando reporte..."):
            time.sleep(1)
            # Aquí llamaríamos a una función de reporte más elaborada
            st.success("Reporte generado:")
            st.write(f"**Reporte para: {audiencia}**")
            st.write(f"- Avance del indicador principal: {indicadores[0].avance:.1f}%")
            st.write(f"- Riesgos activos: {len([r for r in portafolio.riesgos if r['estado'] == 'Activo'])}")
```
