```python
import json
from datetime import datetime

# --- CLASES BASE (Modelos de Datos) ---

class PLADECO:
    def __init__(self, nombre_comuna, periodo, lineamientos):
        self.nombre_comuna = nombre_comuna
        self.periodo = periodo
        self.lineamientos = lineamientos # dict: {"L1": ["O1.1", "O1.2"]}
        self.enfoques = []
        self.marco_seguimiento = {}

class TeoriaDeCambio:
    def __init__(self, nombre, objetivo_pladeco, problema):
        self.nombre = nombre
        self.version = 1.0
        self.objetivo_pladeco = objetivo_pladeco
        self.problema = problema
        self.cadena_causal = {
            "impacto": objetivo_pladeco,
            "resultados": [], # Lista de dicts {"id": "R1", "desc": "...", "tipo": "transformativo"}
            "productos": [],  # Lista de dicts {"id": "P1", "desc": "..."}
            "actividades": [] # Lista de dicts {"id": "A1", "desc": "..."}
        }
        self.supuestos = [] # Lista de dicts {"de": "A1", "a": "P1", "desc": "..."}

class Proyecto:
    def __init__(self, nombre, lider):
        self.nombre = nombre
        self.lider = lider
        self.estado = "Planificado"
        self.progreso = 0
        self.presupuesto_asignado = 0
        self.presupuesto_gastado = 0
        self.actividades_productos = [] # IDs de la TdC

class Portafolio:
    def __init__(self, nombre, tdc_asociada):
        self.nombre = nombre
        self.tdc_asociada = tdc_asociada # Objeto TeoriaDeCambio
        self.proyectos = [] # Lista de objetos Proyecto
        self.riesgos = [] # Lista de dicts {"riesgo": "...", "prob": "Alta", "impacto": "Medio"}

class Indicador:
    def __init__(self, descripcion, resultado_asociado):
        self.descripcion = descripcion
        self.resultado_asociado = resultado_asociado # ID del resultado en la TdC
        self.linea_base = 0
        self.meta = 100
        self.valor_actual = 0
        self.fuente_datos = "Manual" # Puede ser "API: Censo" o "CSV: Encuesta"

    @property
    def avance(self):
        if self.meta == self.linea_base:
            return 100.0 if self.valor_actual >= self.meta else 0.0
        return max(0, min(100, ((self.valor_actual - self.linea_base) / (self.meta - self.linea_base)) * 100))

```
